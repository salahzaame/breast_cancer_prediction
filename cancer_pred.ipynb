```python
#droping the Unamed: 32 column because it containes only NaN values 
data = data.drop("Unnamed: 32", axis=1)
data.isna().sum()
```

```python
data_num = pd.get_dummies(data)
data_num
```

```python
data_num.drop("diagnosis_B", axis=1, inplace=True)
data_num
```

```python
sns.countplot(x='diagnosis_M', data=data_num);
```

```python
# Correlation test :
data_num.corr()
```

# Correlation matrix heatmap

```python
plt.figure(figsize=(25,25))
plt.title('Correlation matrix')
sns.heatmap(data_num.corr(),annot=True , cmap='Greens')
plt.show()
```

```python
#spliting data to X and Y 

x = data_num.drop(["diagnosis_M"], axis=1)
y = data_num["diagnosis_M"]

#spliting data to train-set and test-set

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)


print("training set shape :" , x_train.shape )
print("testing set shape :" , x_test.shape )
```
